```python
adult.head()
```

```python
#Min max normalization of hours-per-week
min_hours = adult["hours-per-week"].min()
max_hours = adult["hours-per-week"].max()
norm_hours = (adult["hours-per-week"]- min_hours)/(max_hours-min_hours)
norm_hours.head()
```

```python
#zscore normalization of age, education and hours-per-week
adult_num = adult[["age", "hours-per-week", "education"]]
zscore = lambda x: (x - x.mean()) / x.std()
adult_std = adult_num.apply(zscore)
adult_std.head()
```

```python
#All three versions of Hours-per-week
norm= pd.DataFrame(adult[["hours-per-week"]])
norm["Min-Max"] = norm_hours
norm["z-score"] = adult_std[["hours-per-week"]]
norm.head(10)
```
